### Changing date type to Month and Year

```python
df['date'] = pd.to_datetime(df['date'])
df['date'].head(15)
```

```python
df['date_year'] = df['date'].apply(lambda x: x.year)
df['date_month'] = df['date'].apply(lambda x: x.month)
```

### Total Number of Houses sold in 2014 and 2015 with their respective Perecntages

```python
total = df.date_year.value_counts()
percentage = df.date_year.value_counts(normalize =True)
print('The total number of houses sold in both years: \n', total)
print('The percentage of houses sold in both years: \n', percentage)
```

### Checking Average Price by year(2014 - 2015) and Month

```python
yearly_price = df.groupby('date_year').mean()['price']
yearly_price
```

```python
plt.figure(figsize=(8,4), dpi = 100)
monthly_price = df.groupby('date_month').mean()['price']
plt.title("Line plot of prices of houses sold per month")
monthly_price.plot()
```

### Using OLS Regression 

```python
independent_variables = df.columns
independent_variables
```

```python
independent_variables = df[['bedrooms', 'bathrooms',
       'floors', 'waterfront', 'view', 'condition', 'grade', 'House_Age']]
dependent_variable = df['price'] 
independent_variables = sm.add_constant(independent_variables)
regression_model = sm.OLS(dependent_variable,independent_variables).fit()
regression_model.summary()
```

```python
#Coefficients that reacts strongly with respect to price
plt.figure(figsize=(10,6), dpi = 100)
regression_model.params.plot(kind='bar')
plt.title("Regressive Coefficients for Price")
```

```python
#Prediction
df['Prediction'] = pd.DataFrame(regression_model.predict(independent_variables))
df['Prediction']
```
